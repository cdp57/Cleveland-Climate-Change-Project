```python
import pandas as pd

print(data['data'])

store = {"date":[], "temp":[], "precipitation":[], "windspeed":[]}
for i in data['data']:
    store["date"].append(i['date'])
    store["temp"].append(i['tavg'])
    store["precipitation"].append(i['prcp'])
    store["windspeed"].append(i['wspd'])

pd.DataFrame(store)
```
